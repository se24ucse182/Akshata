```python
from collections import deque

CAP1 = 4
CAP2 = 3
START = (0, 0)
GOAL = (2, 0)

def get_next_states(state):
    x, y = state
    states = []
    states.append((CAP1, y))
    states.append((x, CAP2))
    states.append((0, y))
    states.append((x, 0))
    transfer = min(x, CAP2 - y)
    states.append((x - transfer, y + transfer))
    transfer = min(y, CAP1 - x)
    states.append((x + transfer, y - transfer))

    return states


# Breadth First Search (BFS)

def bfs():
    visited = set()
    queue = deque([(START, [])])

    while queue:
        state, path = queue.popleft()

        if state in visited:
            continue

        visited.add(state)
        path = path + [state]

        if state == GOAL:
            return path

        for next_state in get_next_states(state):
            if next_state not in visited:
                queue.append((next_state, path))

    return None


# Depth First Search (DFS)

def dfs():
    visited = set()
    stack = [(START, [])]

    while stack:
        state, path = stack.pop()

        if state in visited:
            continue

        visited.add(state)
        path = path + [state]

        if state == GOAL:
            return path

        for next_state in get_next_states(state):
            if next_state not in visited:
                stack.append((next_state, path))

    return None


# Depth Limited Search (DLS)

def dls(state, depth, visited, path):
    visited.add(state)
    path = path + [state]

    if state == GOAL:
        return path

    if depth == 0:
        return None

    for next_state in get_next_states(state):
        if next_state not in visited:
            result = dls(next_state, depth - 1, visited.copy(), path)
            if result:
                return result

    return None


#Iterative Deepening DFS
def iddfs(max_depth=20):
    for depth in range(max_depth):
        result = dls(START, depth, set(), [])
        if result:
            return result
    return None


# Execute All Algorithms

print("BFS Solution:")
print(bfs())
print("\nDFS Solution:")
print(dfs())
print("\nDLS Solution (limit=10):")
print(dls(START, 10, set(), []))
print("\nIDDFS Solution:")
print(iddfs())
```
